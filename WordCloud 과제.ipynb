```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
from wordcloud import WordCloud, STOPWORDS as stopwords
```

```python
pap_text = open("C:/Users/crjh8/Desktop/YS/YBIGTA/2021-1/2.4/과제/Pride_and_Prejudice.txt", "r",
               encoding = 'UTF-8').read()
pap_text
```

```python
stopwords
```

```python
papwc = WordCloud(background_color = "white",
                 collocations = False,
                 stopwords = stopwords)
papwc.generate(pap_text)
```

```python
papwc.words_
```

```python
papwc.to_image()
```

```python
papwc = WordCloud(background_color = 'black',
                 collocations = False,
                 stopwords = stopwords,
                 width = 500, height = 300)
papwc.generate(pap_text)
papwc.to_image()
```

```python
stopwords.update(['said', 'Mr', 's'])

papwc.generate(pap_text)
papwc.to_image()
```

```python
papwc = WordCloud(background_color = 'white',
                 collocations = False,
                 stopwords = stopwords,
                 colormap = 'autumn')

papwc.generate(pap_text)
papwc.to_image()
```
